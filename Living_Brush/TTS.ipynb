```python
import requests
import re
```

```python
def filter_to_hangul_only(text: str) -> str:
    return re.sub(r"[^가-힣\s.,!?]", "", text)
```

```python
def get_tts_url_from_cloud(text: str) -> str:
    CLOUD_RUN_TTS_URL = "https://stt-server-592266359267.asia-northeast3.run.app/tts"
    try:
        response = requests.post(CLOUD_RUN_TTS_URL, json={"text": text}, timeout=10)
        if response.ok:
            return response.json().get("audio")  # 이 URL에는 UUID가 들어감
        else:
            print("⚠️ 상태코드:", response.status_code)
            return None
    except Exception as e:
        print("❌ TTS 요청 실패:", e)
        return None
```

```python
def download_tts_mp3(text: str, save_path: str = "output.mp3"):
    tts_url = get_tts_url_from_cloud(text)
    if not tts_url:
        print("❌ TTS URL을 가져오지 못했습니다.")
        return

    try:
        audio_response = requests.get(tts_url, stream=True)
        if audio_response.ok:
            with open(save_path, "wb") as f:
                for chunk in audio_response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"✅ 저장 완료: {save_path}")
        else:
            print("❌ 음성 파일 다운로드 실패:", audio_response.status_code)
    except Exception as e:
        print("❌ 다운로드 중 오류 발생:", e)
```

```python
text = "안녕하세요! 반딧불이 도우미입니다."


download_tts_mp3(text, "npc_1.mp3")
```

```python
def batch_download_tts(text_list):
    for i, text in enumerate(text_list, start=1):
        filename = f"npc_{i}.mp3"
        download_tts_mp3(text, filename)

# 🎧 음성 리스트
texts = [
    "안녕하세요! 반딧불이 도우미입니다.",
    "무엇을 도와드릴까요~? 반딧불이 도착하면 천천히 말씀해 주세요",
    "다음에 또 불러주세요!",
    "AI 브러쉬 생성을 도와줄께요! 반딧불이 도착하면 천천히 말씀해 주세요",
    "색상 추천을 도와줄께요! 반딧불이 도착하면 천천히 말씀해 주세요",
    "듣고 있어요~! 색상과 원하는 패턴 스타일을 말한 뒤 다시 한번 버튼을 눌러주세요!",
    "듣고 있어요~! 원하는 컬러의 스타일을 말한 뒤 다시 한번 버튼을 눌러주세요!",
    "이제 사용해 보아요!",
    "듣고 있어요! 말이 끝나면 버튼을 다시 눌러주세요!",
    "잠시만요 생각하고 있어요!",
    "돌아오셨군요~ 기다리고 있었어요!",
    "방가워요! 오늘은 어떤 작품일지 너무 궁금해요!",
    "다음에 다시 만나요!"
]

# 📥 실행
batch_download_tts(texts)





```
