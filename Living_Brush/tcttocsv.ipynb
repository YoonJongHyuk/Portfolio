```python
import csv
import io
import re

def tsv_to_csv_string(tsv_text: str) -> str:
    lines = tsv_text.strip().splitlines()
    reader = csv.reader(lines, delimiter='\t')

    # 헤더 처리 (수정/보류 열 제거)
    headers = next(reader)
    headers = [h.strip() for h in headers if h.strip() != "수정/보류"]

    output = io.StringIO()
    writer = csv.writer(output, quoting=csv.QUOTE_MINIMAL)
    writer.writerow(headers)

    for row in reader:
        # 줄바꿈 및 쉼표 제거 + 트림
        cleaned_row = [
            re.sub(r'[\r\n]+', ' ', cell).replace(',', '').strip()
            for cell in row
        ]
        # 딕셔너리로 매핑하고 누락된 필드 보정
        row_dict = dict(zip(headers, cleaned_row))
        filled_row = [row_dict.get(h, '') for h in headers]
        writer.writerow(filled_row)

    return output.getvalue()
```
